```python
# load necessary datasets and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the breast cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
print(data.DESCR) # Display dataset description
```

```python
# Extract the features and target
x = data.data
y = data.target
# Convert to DataFrame for better visualization
df = pd.DataFrame(x, columns=data.feature_names)
df.head()
```

```python

# Check overall DataFrame information
df.info()
```

```python
# no null data
# Standardize the data set: transform the dat set, such that mean is 0 and SD is 1
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
# Convert scaled data to DataFrame for inspection
scaled_df = pd.DataFrame(x_scaled, columns=data.feature_names)
scaled_df.head()
```

```python
# Apply PCA to reduce variables to 2 principal components
# Create 2 PCA components
pca = PCA(n_components=2)
# Fit the scaled data into PCA
x_pca = pca.fit_transform(x_scaled)
# Convert PCA result to DataFrame for better visualization
pca_df = pd.DataFrame(x_pca, columns=['Principal Component 1', 'Principal Component 2'])
print(pca_df.head())
```

```python
# Check the explained variance ratio of each principal comp
print(pca.explained_variance_ratio_)
```

```python
# Add target labels to the PCA DataFrame
pca_df['Target'] = y  # y = data.target

# Define colors and labels
colors = ['red', 'blue']
labels = data.target_names  # ['malignant', 'benign']

# Scatter plot for each class
for i, label in enumerate(labels):
    plt.scatter(
        pca_df.loc[pca_df['Target'] == i, 'Principal Component 1'],  # X-axis: PC1
        pca_df.loc[pca_df['Target'] == i, 'Principal Component 2'],  # Y-axis: PC2
        color=colors[i],                            # Assign color
        label=label,                                # Legend label
        alpha=0.5                                   # Transparency
    )

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Breast Cancer Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Logistic regression for prediction; x_pca contains the principal components and y contains the target labels
# split the data in 30% test and 70% train, ensure reproducibility with random_state
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.3, random_state=42)

model = LogisticRegression()
model.fit(x_train, y_train)

accuracy = model.score(x_test, y_test)
print(f"Logistic Regression Accuracy on PCA components: {accuracy:.2f}") 
# 97% accuracy shows a well trained model predicting breast cancer malignancy using PCA components
```
